```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\harka\Datasets\weatherHistory.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.columns
```

```python
#Changing tha data type of 'Formatted Date' from to datetime
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'], utc=True)
data['Formatted Date']
```

```python
data.describe()
```

```python
#setting the 'Formatted Date' as index
data = data.set_index('Formatted Date')
data.head()
```

```python
#Coverting the hourly data to monthly data and keeping only the 'Apparent Temperature (C)' and 'Humidity' columns
data_columns = ['Apparent Temperature (C)', 'Humidity']data_monthly = data[data_columns].resample('MS').mean()
data_monthly.tail()
```

```python
data_montly['Apparent Temperature (C)']
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=data_monthly, marker='o')
plt.show()
```

Obsevation:
From the graph we we can the humidity 10 years is constant and apparent temperature is also same for all years except for 2012. In 2012 apparent temperature falls down a little bit.
